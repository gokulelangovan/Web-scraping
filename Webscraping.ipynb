```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
services_type = soup.find(id="services")
country_info = services_type.find(class_="services-country-grids")
print(country_info.prettify())
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
services_type = soup.find(id="services")
topic_info = services_type.find(class_="services-info-topic")
print(topic_info.prettify())
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
services_type = soup.find(id="services")
topic_detail = services_type.find_all(class_="topic-detail")
inside_engg_and_tech = topic_detail[3]
print(inside_engg_and_tech.prettify())
```
